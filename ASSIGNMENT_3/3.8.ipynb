```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver11=webdriver.Chrome(r"chromedriver.exe")  
```

```python
driver11.get("https://www.youtube.com/watch?v=T936yTchDck")
```

```python
for _ in range(1000):
    driver11.execute_script("window.scrollBy(0,1000)")
```

```python
comment11=[]
com11_tags=driver11.find_elements(By.XPATH,'//yt-formatted-string[@class="style-scope ytd-comment-renderer"]')
for i in com11_tags[0:500]:
    comment11.append(i.text)
comment11 
```

```python
print(len(comment11))
```

```python
upvote11=[]
uv11_tags=driver11.find_elements(By.XPATH,'//span[@id="vote-count-middle"]')
for i in uv11_tags[0:500]:
    upvote11.append(i.text)
upvote11 
```

```python
print(len(upvote11))
```

```python
time11=[]
time11_tags=driver11.find_elements(By.XPATH,'//yt-formatted-string[@class="published-time-text style-scope ytd-comment-renderer"]')
for i in time11_tags[0:500]:
    time11.append(i.text)
time11 
```

```python
print(len(time11))
```

```python
df11=pd.DataFrame({'Comment':comment11,'Comment Upvote':upvote11,'Time when comment was posted':time11})
df11
```
