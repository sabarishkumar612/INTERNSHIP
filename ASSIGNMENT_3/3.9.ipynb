```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver12=webdriver.Chrome(r"chromedriver.exe")
```

```python
driver12.get("https://www.hostelworld.com/")
```

```python
lon12=driver12.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input')
lon12.send_keys('London') 
```

```python
search12=driver12.find_element(By.XPATH,'//div[@class="label"]')
search12.click() 
```

```python
search122=driver12.find_element(By.XPATH,'//button[@class="button primary large"]')
search122.click() 
```

```python
url12=[]
url122=driver12.find_elements(By.XPATH,'//h2[@class="title title-6"]/a')
for i in url122:
    url12.append(i.get_attribute('href'))
    time.sleep(3)
url12
```

```python
pd12=[]
for i in url12:
    driver12.get(i)
    time.sleep(3)
    try:
        prodes12=driver12.find_element(By.XPATH,'//div[@class="description-container"]/div/div[2]')
        pd12.append(prodes12.text)
    except NoSuchElementException:
        pd12.append('-')
pd12
```

```python
print(len(pd12))
```

```python
dis12=[]
dis12tags=driver12.find_elements(By.XPATH,'//span[@class="description"]')
for i in dis12tags:
    dis12.append(i.text.split('-')[1])
dis12
```

```python
print(len(dis12))
```

```python
name12=[]
name12tags=driver12.find_elements(By.XPATH,'//h2[@class="title title-6"]')
for i in name12tags:
    name12.append(i.text)
name12
```

```python
rate12=[]
rate12_tags=driver12.find_elements(By.XPATH,'//*[@class="score orange big" or @class="score gray big"]')
for i in rate12_tags:
    rate12.append(i.text)
rate12
```

```python
review12=[]
review12_tags=driver12.find_elements(By.XPATH,'//div[@class="reviews"]')
for i in review12_tags:
    review12.append(i.text.split(' ')[0])
review12
```

```python
oreview12=[]
oreview12_tags=driver12.find_elements(By.XPATH,'//div[@class="keyword"]')
for i in oreview12_tags:
    oreview12.append(i.text)
oreview12
```

```python
ore12=oreview12[1:]
ore12
```

```python
print(len(ore12))
```

```python
pp12=[]
pp12_tags=driver12.find_elements(By.XPATH,'//div[@class="price-col"][1]')
for i in pp12_tags:
    pp12.append(i.text)
pp12
```

```python
print(len(pp12))
```

```python
dp12=[]
dp12_tags=driver12.find_elements(By.XPATH,'//div[@class="price-col"][2]')
for i in dp12_tags:
    dp12.append(i.text)
dp12
```

```python
df12=pd.DataFrame({'Hostel Name':name12,'Distance From City Centre':dis12,'Ratings':rate12,'Total Reviews':review12,'Overall Review':ore12,'Privates from Price':pp12,'Dorms from Price':dp12,'Property Description':pd12,'Hostel URL':url12})
df12
```
