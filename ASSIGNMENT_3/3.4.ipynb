```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver6=webdriver.Chrome(r"chromedriver.exe")   
```

```python
driver6.get("http://www.flipkart.com/")    
```

```python
p1=driver6.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")
p1.send_keys('smartphone')  
```

```python
search10=driver6.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search10.click() 
```

```python
url1=[]
u1=driver6.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in u1:
    url1.append(i.get_attribute('href'))
    time.sleep(3)
url1
```

```python
print(len(url1))
```

```python
product1=[]
for i in url1:
    driver6.get(i)
    time.sleep(3)
    try:
        name1=driver6.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        product1.append(name1.text)
    except NoSuchElementException:
        product1.append('-')
product1
```

```python
print(len(product1))
```

```python
ed1=[]
for i in url1:
    driver6.get(i)
    time.sleep(3)
    try:
        dele1=driver6.find_element(By.XPATH,'//span[@class="_1TPvTK"]')
        ed1.append(dele1.text)
    except NoSuchElementException:
        ed1.append('-')
ed1
```

```python
price1=[]
for i in url1:
    driver6.get(i)
    time.sleep(3)
    try:
        pri1=driver6.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        price1.append(pri1.text)
    except NoSuchElementException:
        price1.append('-')
price1
```

```python
c1=[]
for i in url1:
    driver6.get(i)
    time.sleep(3)
    try:
        c2=driver6.find_element(By.XPATH,'//div[@class="_3k-BhJ"][1]/table/tbody/tr[4]/td[2]')
        c1.append(c2.text)
    except NoSuchElementException:
        c1.append('-')
c1
```

```python
rom1=[]
for i in url1:
    driver6.get(i)
    time.sleep(3)
    try:
        r1=driver6.find_element(By.XPATH,'//li[@class="_21Ahn-"][1]')
        rom1.append(r1.text)
    except NoSuchElementException:
        rom1.append('-')
rom1
```
