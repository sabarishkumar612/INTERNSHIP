```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")  
```

```python
driver.get("https://www.amazon.in/")        
```

```python
a=input("Enter the product to be explored")
```

```python
p=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
p.send_keys(a)   
```

```python
search=driver.find_element(By.XPATH,"/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
search.click()      
```

# PROBLEM 2

```python
purl=[]
for page in range (0,3):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        purl.append(i.get_attribute('href'))
    button=driver.find_element(By.XPATH,'//span[@class="s-pagination-strip"]/a[3]')
    button.click()
    time.sleep(3)
purl
```

```python
print(len(purl))
```

```python
product=[]
for i in purl:
    driver.get(i)
    time.sleep(3)
    try:
        name=driver.find_element(By.XPATH,'//h1[@id="title"]')
        product.append(name.text)
    except NoSuchElementException:
        product.append('-')
product
```

```python
print(len(product))
```

```python
ed=[]
for i in purl:
    driver.get(i)
    time.sleep(3)
    try:
        dele=driver.find_element(By.XPATH,'//div[@class="a-spacing-base"][1]')
        ed.append(dele.text)
    except NoSuchElementException:
        ed.append('-')
ed
```

```python
print(len(ed))
```

```python
price=[]
for i in purl:
    driver.get(i)
    time.sleep(3)
    try:
        pri=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none aok-align-center"]/span[2]')
        price.append(pri.text)
    except NoSuchElementException:
        price.append('-')
price
```

```python
print(len(price))
```

```python
rere=[]
for i in purl:
    driver.get(i)
    time.sleep(3)
    try:
        rr=driver.find_element(By.XPATH,'//div[@id="iconfarmv2_feature_div"]/div[2]/div/div/div/div[2]/div/ol/li[2]/div/span/div[2]/a')
        rere.append(rr.text)
    except NoSuchElementException:
        rere.append('-')
rere
```

```python
brand=[]
for i in purl:
    driver.get(i)
    time.sleep(3)
    try:
        br=driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]/td[2]')
        brand.append(br.text)
    except NoSuchElementException:
        brand.append('-')
brand
```

```python
df=pd.DataFrame({'PRODUCT NAME':product,'BRAND':brand,'PRICE':price,'EXPECTED DELIVERY':ed,'RETURN/EXCHANGE':rere,'PRODUCT URL':purl})
df
```
